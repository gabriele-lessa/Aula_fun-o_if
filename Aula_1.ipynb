```python
# 1 - criar uma lista com nome de 5 pets
lista_pets = ['gato',
            'cachorro',
            'peixe',
            'cobra',
            'papagaio']

# 2 - criar uma função para exibir todos os pets
def ver_pets():
  print(f'os pets que temos são: {lista_pets}')

# 3 - criar uma função para cadastrar um novo pet
def adicionar_pets():
  novo_pet = input('informe qual o seu pet: ')
  lista_pets.append(novo_pet)

# 4 - criar um menu de opções para ver ou cadastrar pets
print('-----Cadrasto dos Pets-----')
print('escolha a opção: ')
print('1- ver pet')
print('2- cadastrar novo pet')
opcao = int(input('Informe a opcao desejada'))

#Comando IF para condicionar
if opcao ==1:
  #o que ele deve executar
  ver_pets()
elif opcao ==2:
    adicionar_pets()
    ver_pets()
else:
  print('opcao não encontrada')
```
